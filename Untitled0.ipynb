** Hard-Coding Variables**

```python
import math
import matplotlib.pyplot as plt

def quadratic_solver(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return root1, root2
    elif discriminant == 0:
        root = -b / (2*a)
        return root, root
    else:
        return None

def plot_weather_model(a, b, c):
    x = range(-10, 11)
    y = [a * i**2 + b * i + c for i in x]
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f"{a}x² + {b}x + {c}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title("Weather Model - Hardcoded Variables")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid(True)
    plt.show()

# Hardcoded coefficients
a, b, c = 4, -7, 3
solutions = quadratic_solver(a, b, c)
print("Solutions:", solutions)
plot_weather_model(a, b, c)
```

** Reading from a File (Single Set)**

```python
import math
import matplotlib.pyplot as plt

def quadratic_solver(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return root1, root2
    elif discriminant == 0:
        root = -b / (2*a)
        return root, root
    else:
        return None

def plot_weather_model(a, b, c):
    x = range(-10, 11)
    y = [a * i**2 + b * i + c for i in x]
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f"{a}x² + {b}x + {c}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title("Weather Model - File Input")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid(True)
    plt.show()

filename = input("Enter the file path: ")
try:
    with open(filename, 'r') as file:
        for line in file:
            coefficients = list(map(float, line.split()))
            if len(coefficients) == 3:
                a, b, c = coefficients
                solutions = quadratic_solver(a, b, c)
                print(f"Coefficients: {coefficients} -> Solutions: {solutions}")
                plot_weather_model(a, b, c)
            else:
                print(f"Invalid line skipped: {line}")
except FileNotFoundError:
    print("File not found. Please provide a valid path.")
except ValueError:
    print("Error reading coefficients. Ensure the file contains valid numbers.")
```

**Multiple Sets of Inputs**

```python
import math
import matplotlib.pyplot as plt

def quadratic_solver(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return root1, root2
    elif discriminant == 0:
        root = -b / (2*a)
        return root, root
    else:
        return None

def plot_weather_model(a, b, c):
    x = range(-10, 11)
    y = [a * i**2 + b * i + c for i in x]
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f"{a}x² + {b}x + {c}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title("Weather Model - Multiple Inputs")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid(True)
    plt.show()

print("Enter multiple sets of coefficients (a, b, c) separated by commas. Example: 1 2 1, 4 -7 3")
inputs = input("Enter coefficients: ").split(',')
for coeffs in inputs:
    a, b, c = map(float, coeffs.split())
    solutions = quadratic_solver(a, b, c)
    print(f"Solutions for {a}x² + {b}x + {c}: {solutions}")
    plot_weather_model(a, b, c)
```
